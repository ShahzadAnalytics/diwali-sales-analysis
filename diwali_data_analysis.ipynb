```python
# Top 10 states by Orders
sales_order = df.groupby('State', as_index=False)['Orders'].sum() \
                .sort_values(by='Orders', ascending=False).head(10)
# Set figure size
sns.set(rc={'figure.figsize':(15,5)})
# Plot
sns.barplot(data=sales_order, x='State', y='Orders', palette='deep')
```

Matrial status

```python
ax=sns.countplot(data=df,x='Shadi', palette='deep')
# Set figure size
sns.set(rc={'figure.figsize':(2,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# Group by Shadi and Gender, sum Amount
shadi_sales = df.groupby(['Shadi', 'Gender'], as_index=False)['Amount'].sum() \
                .sort_values(by='Amount', ascending=False)

# Plot
plt.figure(figsize=(8,5))
ax = sns.barplot(data=shadi_sales, x='Shadi', y='Amount', hue='Gender', palette='pastel')
# Add labels on top of bars
for bar in ax.containers:
    ax.bar_label(bar, label_type='edge', fontsize=10)

plt.title("Total Sales by Marital Status and Gender", fontsize=14)
plt.xlabel("Marital Status")
plt.ylabel("Total Sales Amount")
plt.show()
```

 conclusion
  
#from the above graph we can see that most of the buyers are married (womens) and they  have high purchaising power  

Occupation 

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation', palette='deep')
# Set figure size
for bar in ax.containers:
    ax.bar_label(bar)
```

```python

sales_order = df.groupby('Occupation', as_index=False)['Amount'].sum() \
                .sort_values(by='Amount', ascending=False).head(10)
# Set figure size
sns.set(rc={'figure.figsize':(15,5)})
# Plot
sns.barplot(data=sales_order, x='Occupation', y='Amount', palette='deep')
```

From the above graph we see that most of the buyers are  working in IT,healthcare,Avition,Banking

Product_Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Product_Category', palette='deep')
# Set figure size
for bar in ax.containers:
    ax.bar_label(bar)
```

```python

sales_order = df.groupby('Product_Category', as_index=False)['Amount'].sum() \
                .sort_values(by='Amount', ascending=False).head(10)
# Set figure size
sns.set(rc={'figure.figsize':(15,5)})
# Plot
sns.barplot(data=sales_order, x='Product_Category', y='Amount', palette='deep')
```

from the above graph we can see that most of the sold product are from food,cloths and electronic categrory

```python
# Group by Product_ID and sum Orders
top_products = df.groupby('Product_ID', as_index=False)['Orders'].sum() \
                 .sort_values(by='Orders', ascending=False).head(10)

# Set figure size
sns.set(rc={'figure.figsize': (15,5)})

# Plot
ax = sns.barplot(data=top_products, x='Product_ID', y='Orders', palette='pastel')

# Add labels on top of bars
for bar in ax.containers:
    ax.bar_label(bar, label_type='edge', fontsize=10)
```
